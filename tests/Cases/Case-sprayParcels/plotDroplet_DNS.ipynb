```python
# Plot the droplet evaporation 

import matplotlib.pyplot as plt
import numpy as np

```

```python
# ==============================================================================
# Ethanol Droplet Evaporation       
# ==============================================================================


# Load data
data_EE = np.genfromtxt('dropletDiameter_DNS_Ethanol_EulerExplicit.dat')
data_EE_OF = np.genfromtxt('dropletDiameter_DNS_Ethanol_EulerExplicit_OFLiquid.dat')
data_CN = np.genfromtxt('dropletDiameter_DNS_Ethanol_CrankNicolson.dat')
expData = np.genfromtxt('expEthanolDroplet.dat',delimiter=',',skip_header=2)
# Results from Matlab by Marvin Sontheimer
simMS = np.genfromtxt('dropletDiameter_Ethanol_Sontheimer.dat',delimiter=',',skip_header=2)
# Initial droplet diameter
d0=6E-4


fSize = 22
lw = 2.5
plt.rc('text', usetex=True)
plt.rc('text.latex',preamble=r'\usepackage{amsmath}')
plt.rc('font', family='serif')
plt.rc('font', size=fSize)
plt.rc('figure',figsize=[7,6])



plt.figure()
plt.plot(data_EE[:,0]/np.power(d0*1000,2),np.power(data_EE[:,1]/d0,2),label='Sim. EE')
plt.plot(data_EE_OF[:,0]/np.power(d0*1000,2),np.power(data_EE_OF[:,1]/d0,2),label='Sim. EE - OF')
plt.plot(data_CN[:,0]/np.power(d0*1000,2),np.power(data_CN[:,1]/d0,2),label='Sim. CN',ls='--')
plt.plot(simMS[:,0]/np.power(d0*1000,2),np.power(simMS[:,1]/d0,2),label='Matlab',ls='-')
plt.plot(expData[:,0]/np.power(d0*1000,2),expData[:,1],ls='None',marker='o',label='Exp. (Ma et al. 2013)')
plt.xlabel(r'$(t/d_0^2) [s/mm^2]$')
plt.ylabel(r'$(d/d_0)^2$')
plt.legend(fontsize='xx-small')
plt.savefig('ethanolDropletEvap.png',dpi=300,format='png')


plt.figure()
plt.plot(data_EE[:,0]/np.power(d0*1000,2),data_EE[:,2],label='Sim. EE')
plt.plot(data_EE_OF[:,0]/np.power(d0*1000,2),data_EE_OF[:,2],label='Sim. EE - OF')
plt.plot(data_CN[:,0]/np.power(d0*1000,2),data_CN[:,2],label='Sim. CN')
plt.plot(simMS[:,0]/np.power(d0*1000,2),simMS[:,2],label='Matlab')
plt.xlabel(r'$(t/d_0^2) [s/mm^2]$')
plt.ylabel(r'$T [K]$')
plt.ylim([290,315])
plt.legend()
plt.savefig('ethanolDropletEvap_T.png',dpi=300,format='png')    
plt.legend(fontsize='xx-small')
```

```python
# ==============================================================================
# n-Heptane Droplet Evaporation
# ==============================================================================


# Load data
data_EE = np.genfromtxt('dropletDiameter_nHeptane.dat')
simData_Wong_Case1 = np.genfromtxt('YangWong2001_555K.dat',delimiter=' ',skip_header=2)
expData = np.genfromtxt('expNHeptaneDroplet.dat',delimiter=',',skip_header=2)
simMS = np.genfromtxt('dropletDiameter_nHeptane_Sontheimer.dat',delimiter=',',skip_header=2)
# Initial droplet diameter
d0=7E-4


fSize = 22
lw = 2.5
plt.rc('text', usetex=True)
plt.rc('text.latex',preamble=r'\usepackage{amsmath}')
plt.rc('font', family='serif')
plt.rc('font', size=fSize)
plt.rc('figure',figsize=[7,6])



plt.figure()
plt.plot(data_EE[:,0]/np.power(d0*1000,2),np.power(data_EE[:,1]/d0,2),label='Sim.')
plt.plot(simData_Wong_Case1[:,0],simData_Wong_Case1[:,1],label='Yang \& Wong')
plt.plot(simMS[:,0]/np.power(d0*1000,2),np.power(simMS[:,1]/d0,2),label='Matlab')
plt.plot(expData[:,0],expData[:,1],ls='None',marker='o',label='Exp. (Nomura et al.)')
plt.xlabel(r'$(t/d_0^2) [s/mm^2]$')
plt.ylabel(r'$(d/d_0)^2$')
plt.legend()


plt.figure()
plt.plot(data_EE[:,0]/np.power(d0*1000,2),data_EE[:,2],label='Sim.')
plt.plot(simMS[:,0]/np.power(d0*1000,2),simMS[:,2],label='Matlab')
plt.ylim([290,380])
plt.xlabel(r'$(t/d_0^2) [s/mm^2]$')
plt.ylabel(r'$T [K]$')
plt.legend()
```
